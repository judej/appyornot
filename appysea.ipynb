```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pie Chart
labels = ['Antibiotics', 'Appendectomy']
sizes = [47, 53]
colors = ['#ff9999','#66b3ff']
 
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Treatment Assignment')
plt.show()

# Horizontal Bar Chart
complications_antibiotics = 8.1
complications_appendectomy = 3.5

data = {'Treatment Group': ['Antibiotics', 'Appendectomy'], 
        'Complications Rate': [complications_antibiotics, complications_appendectomy]}

sns.set_style("whitegrid")
sns.barplot(x="Complications Rate", y="Treatment Group", data=data, palette=colors)
plt.title('Complication Rates by Treatment Group')
plt.show()

# Vertical Bar Chart
serious_adverse_events_antibiotics = 4.0
serious_adverse_events_appendectomy = 3.0

data = {'Treatment Group': ['Antibiotics', 'Appendectomy'], 
        'Serious Adverse Events Rate': [serious_adverse_events_antibiotics, serious_adverse_events_appendectomy]}

sns.set_style("whitegrid")
sns.barplot(x="Treatment Group", y="Serious Adverse Events Rate", data=data, palette=colors)
plt.title('Serious Adverse Event Rates by Treatment Group')
plt.show()
```
